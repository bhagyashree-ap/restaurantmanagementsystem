```python
#restaurantmanagementsystem
import matplotlib.pyplot as plotgraph
import tkinter as tk
hotel={1:['1', 'chocolate Dessert','40'], 2:['2', 'Cheesecake','70'],
       3:['3', 'vanilla cake','60']}   #tablenumber,order,bill

def add():
    n=[]
    TN=input("Enter table number: ")
    Order=input("Enter order: ")
    Bill=input("Enter total bill: ")
    i=hotel.keys()
    if TN not in i:
        n.append(TN)
        n.append(Order)
        n.append(Bill)
        hotel[TN]=n
        print(hotel)
    else: 
        print("Table number already exists!")
        
def replace_order():
    n=[]
    TN=int(input("enter table number: "))
    Order=input("Enter order: ")
    Bill=input("enter total bill: ")
    i=hotel.keys()
    if TN in i:
        n.append(TN)
        n.append(Order)
        n.append(Bill)
        hotel[BN]=n
        print(hotel)
    else:
        print("Table number does not exist!")
        
def delete():
    BN=int(input("Enter table number: "))
    i=hotel.keys()
    if TN in i:
        del hotel[TN]
        print(hotel)
    else:
        print("Cannot find the table number!")
    
def order():
    global order
    global count
    order=[]
    count=[]
    for i in hotel.values():
        if i[1] not in order:
            order.append(i[1])
    l=list(hotel.values())
    for temp in order:
        c=0
        for i in range(len(hotel)):
            if temp==l[i][1]:
                c=c+1
        count.append(c)
    print(count)
    plotgraph.bar(order,count,width=0.5, align='center')
    plotgraph.show

def bill():
    global b
    global count_b
    b=[]
    count_b=[]
    for i in hotel.values():
        if i[2] not in b:
            b.append(i[2])
    lis=list(hotel.values())
    for temp in b:
        c=0
        for i in range(len(hotel)):
            if temp==lis[i][2]:
                c=c+1
        count_b.append(c)
    print(count_b)
    plotgraph.bar(b,count_b,width=0.5, align='center')
    plotgraph.show
    
root=tk.Tk()
root.title('Restaurant Management System')
frame=tk.Frame(root)
frame.pack()
button=tk.Button(frame, text='Add Order', font=('arial', 15, 'bold'), width=40, command=add)
button.pack()
button=tk.Button(frame, text='Remove Order', font=('arial', 15, 'bold'), width=40, command=delete)
button.pack()
button=tk.Button(frame, text='Replace Order', font=('arial', 15, 'bold'), width=40, command=replace_order)
button.pack()
button=tk.Button(frame, text='Order Graph', font=('arial', 15, 'bold'), width=40, command=order)
button.pack()
button=tk.Button(frame, text='Bill Graph', font=('arial', 15, 'bold'), width=40, command=bill)
button.pack()
button=tk.Button(frame, text='Quit', font=('arial', 15, 'bold'), width=40, command=root.destroy)
button.pack()
root.mainloop()
```
